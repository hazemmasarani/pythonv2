# Loops in Python


### Objectives
- Understand the concept and application of loops in Python.
- Learn the differences and use cases of `while` and `for` loops.
- Explore how to control loops using `break` and `continue`.
- Investigate the concept and utility of nested loops.

### 1. Understanding While and For Loops
Loops enable the execution of a code block multiple times until a condition is met.

#### While Loop
A `while` loop runs as long as a condition is True.

Example:
```python
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```


This code increments and prints a counter until it reaches 5.

#### For Loop
A `for` loop iterates over a sequence of elements.

Example:
```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```
This loop prints each fruit in the `fruits` list.

### 2. Controlling Loops with Break and Continue
`break` and `continue` control the flow within loops.

#### Break Statement
The `break` statement exits the loop prematurely.

Example:
```python
count = 0
while count < 5:
    print("Count:", count)
    count += 1
    if count == 3:
        break
```
This loop stops when `count` reaches 3.

#### Continue Statement
The `continue` statement skips the current iteration and proceeds to the next one.

Example:
```python
for num in range(10):
    if num % 2 == 0:
        continue
    print(num)
```
This loop prints only odd numbers from 0 to 9.

### 3. Nesting Loops
Nested loops are loops within loops, allowing for complex iterations.

Example:
```python
for i in range(3):
    for j in range(2):
        print(i, j)
```
This code prints pairs of numbers, demonstrating two levels of looping.

### Summary
- Loops in Python are used for repetitive tasks.
- `while` loops run based on a condition, and `for` loops iterate over sequences.
- `break` and `continue` control the loop execution.
- Nested loops facilitate multi-dimensional iterations.
